```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import numpy as np

# Generate time values
time = np.linspace(0, 2*np.pi, 10)

# Generate sinus time series
series1 = np.sin(time) + np.cos(time) * 0.5
series2 = np.sin(time) * 2 + np.cos(time) * 0.2
series3 = np.sin(time) * 3 + np.cos(time) * 0.7
series4 = np.sin(time) * 1 + np.cos(time)

series = np.array([series1, series2, series3, series4])

series_modified = series.copy()
series_modified[:, -1] -= 3

# Plot the time series
fig, ax = plt.subplots(2, 1, figsize=(10, 10))
ax[0].plot(time, series.T)
ax[0].set_title("Original time series")
ax[1].plot(time, series_modified.T)
ax[1].set_title("Time series with last value subtracted by 3")
plt.tight_layout()
plt.show()
```
