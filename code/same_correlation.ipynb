```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'DejaVu Serif'
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['lines.markersize'] = 12
plt.rcParams['xtick.labelsize'] = 24
plt.rcParams['ytick.labelsize'] = 24
plt.rcParams['legend.fontsize'] = 24
plt.rcParams['axes.titlesize'] = 36
plt.rcParams['axes.labelsize'] = 24
```

```python
import numpy as np

# Generate time values
time = np.linspace(0, 2*np.pi, 10)

# Generate sinus time series
series1 = np.sin(time) + np.cos(time) * 0.5
series2 = np.sin(time) * 2 + np.cos(time) * 0.2
series3 = np.sin(time) * 3 + np.cos(time) * 0.7

series = np.array([series1, series2, series3])

series_modified = series.copy()
series_modified[:, -1] -= 3

# Plot the time series
plt.figure(figsize=(12, 8))
plt.plot(time, series[0], label='Function 1', color='red', alpha=0.9, linestyle='--')
plt.plot(time, series[1], label='Function 2', color='green', alpha=0.9, linestyle='--')
plt.plot(time, series[2], label='Function 3', color='blue', alpha=0.9, linestyle='--')
plt.legend()
plt.savefig('../paper/figures/Original_functions.eps', format='eps')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.plot(time, series_modified[0], label='Function 1 modified', color='red', alpha=0.9, linestyle='--')
plt.plot(time, series_modified[1], label='Function 2 modified', color='green', alpha=0.9, linestyle='--')
plt.plot(time, series_modified[2], label='Function 3 modified', color='blue', alpha=0.9, linestyle='--')
plt.legend()
plt.savefig('../paper/figures/Modified_functions.eps', format='eps')
plt.show()
```
